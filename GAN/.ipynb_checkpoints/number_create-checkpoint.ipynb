```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits

from keras.models import Model
from keras.layers import Dense, Input
```

```python
dataset = load_digits()
x = dataset.data
y = dataset.target
print(x.shape, y.shape)
x3 = x[y==3]
plt.imshow(x3[0].reshape([8, 8]))
print(x3.shape)
```

```python
def create_generator():
    g_in = Input(shape=8)
    model = Dense(128, activation='relu')(g_in)
    g_out = Dense(64)(model)
    return Model(g_in, g_out)

def create_discriminator():
    d_in = Input(shape=64)
    model = Dense(32, activation='relu')(d_in)
    d_out = Dense(1, activation='sigmoid')(model)
    return Model(d_in, d_out)
```

```python
np.random.seed(1)
right = np.ones([183, 1])
wrong = np.zeros([183, 1])

discriminator = create_discriminator()
discriminator.compile(optimizer='adam', loss='mse')
generator = create_generator()
g_in = Input(shape=8)
g_out = generator(g_in)
discriminator.trainable = False
d_out = discriminator(g_out)
model = Model(g_in, d_out)
model.compile(optimizer='adam', loss='mse')

EPOCHS = 1000
seed = np.zeros([183, 8])

for epoch in range(EPOCHS):
    seed = np.random.randn(183, 8)
    fake = generator.predict(seed)
    real_error = discriminator.train_on_batch(x3, right)
    fake_error = discriminator.train_on_batch(fake, wrong)
    
    seed = np.random.randn(183, 8)
    model.train_on_batch(seed, right)
    
    if epoch % 100 == 0:
        plt.imshow(fake[0].reshape([8, 8]))
        plt.show()
plt.imshow(fake[0].reshape([8, 8]))
plt.show()
```

```python
seed = np.random.randn(1, 8)
number = generator.predict(seed)
plt.imshow(number.reshape([8, 8]))
plt.show()
```
